```python
#Import Librarairs
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the data from source
df = pd.read_csv(r'C:\Users\leevc\Downloads\movies.csv')
```

```python
#Look into the data
df
```

```python
#Check for any missing data
df.isnull().sum()
```

```python
df.dropna(inplace=True)

df.isnull().sum()
```

```python
#To check data types of each column
df.dtypes
```

```python
#To arrange with the highest revenue value in descending order
df.sort_values(by=['revenue'], inplace = False, ascending = False)
```

```python
#To plot using seaborn to see the correlation between revenue and budget
sns.regplot(x="revenue", y="budget", data=df, scatter_kws={'color':'red'}, line_kws={'color': 'blue'})
```

```python
#To check on the actual correlation using Pearson method

df.corr(method='pearson')
```

```python
#To show the correlation matrix where Budget and Gross has the highest correlation
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show
```

```python
#Unstack to see in better way
correlation_mat = df.corr(method='pearson')
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
#Sort to see and compare in a quicker way
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#To see correlation > 0.1 only
high_corr = sorted_pairs[(sorted_pairs)>0.1]
high_corr
```
